```python
import re
```

```python
txt = 'El 10 de agosto de 2025, María Pérez envió un correo electrónico a la dirección maria.perez@email.com solicitando información sobre el producto X-200. \nSu número de teléfono es +34 612-345-678. Además, visitó la web https://www.ejemplo.com para consultar precios y disponibilidad.\n\nEl servidor registró el acceso desde la IP 192.168.1.15 a las 14:23:56, y se detectó un error en el log con código ERR500. En Twitter, el usuario @DataAnalyst \ncomentó: "Los modelos de IA son el futuro #MachineLearning #IA2025". El totaal de la compra fue de $199.99, y el cliente pagó con tarjeta VISA terminada en 1234.\n\nPor otro lado, el DNI de Juan García es 12345678Z y su código postal 28080. Para más información, puede llamar al 911 234 567 o visitar www.informacion.es. \nRecordamos que las feechas límite para inscripción son 2025-09-30 y 2025-10-15.'

print(txt)
```

### **1. Extraer todas las direcciones de correo electrónico**

```python
pattern = r'\S+@\S+'

re.findall(pattern, txt)
```

**2.	Extraer todas las URLs (con o sin protocolo)**

```python
url = r'(?:\S+)?www+\.\S+'

re.findall(url, txt)
```

 **3.	Validar y extraer números de teléfono en formato internacional y nacional.**

```python
num_tel = r"(?:\b\d{3}\s\d{3}\s\d{3}\b|\+\d{1,3}\s\d{3}(?:[-\s]\d{3}){2}\b)"

re.findall(num_tel, txt)
```

 **4.	Extraer fechas en formato yyyy-MM-dd.**

```python
date = r'\d{4}-\d{2}-\d{2}'

re.findall(date, txt)
```

### **5.	Extraer códigos postales de 5 dígitos.**

```python
CP = r'\d{5}\b'

re.findall(CP, txt)
```

### **6.	Extraer DNI (8 dígitos seguidos de letra mayúscula).**

```python
DNI = r'\d{8}+\w'

re.findall(DNI, txt)
```

 **7.	Extraer menciones de Twitter.**

```python
mencion = r'\s@\w+'

re.findall(mencion, txt)
```

 **8.	Extraer hashtags.**

```python
hashtag = r'#\w+'

re.findall(hashtag, txt)
```

 **9.	Extraer direcciones IP.**

```python
IP = r'\d{3}\.\d{3}\.\d{1}\.\d{2}'

re.findall(IP, txt)
```

 **10.	Extraer importes monetarios (con $ y decimales).**


```python
importes = r'\$\d+\.\d+'

re.findall(importes, txt)
```
